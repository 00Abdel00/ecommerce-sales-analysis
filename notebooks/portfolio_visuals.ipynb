```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the data
df = pd.read_csv(r"C:\Users\sypho\portfolio\PROJETS\ecom\ecommerce-sales-analysis\data\Amazon_Sale_Report_Cleaned.csv")
# Reset Index if error
df.reset_index(inplace=True)
```

# üìÅ Portfolio Visuals ‚Äì Amazon Sales EDA
Clean and annotated charts selected from the full analysis.

### üìä Monthly Revenue Trend (Millions IDR)

```python
# Set Date as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Compute monthly total revenue and convert to millions
monthly_revenue = df['Amount (IDR)'].resample('M').sum()
monthly_revenue_million = monthly_revenue / 1_000_000

# Remove March (month 3)
monthly_revenue_million = monthly_revenue_million[monthly_revenue_million.index.month != 3]

# Format the index for display (e.g., "Apr 2022")
monthly_revenue_million.index = monthly_revenue_million.index.strftime('%b %Y')

# Plot
plt.figure(figsize=(10, 5))
monthly_revenue_million.plot(kind='bar', color='skyblue')

# Titles and labels
plt.title("Monthly Revenue Trend (in Millions IDR)", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Revenue (Million IDR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Reset Index
df.reset_index(inplace=True)

# Save as image
plt.savefig("../visualizations/monthly_revenue.png", dpi=300)

plt.show()
```

This bar chart shows the total monthly revenue from orders between April and June.  
Revenue decreased consistently over the three-month period, indicating a potential drop in demand, seasonality, or platform-specific factors.

- **April** had the highest revenue.
- **June** closed with the lowest performance.
- Currency is in **Indonesian Rupiah (IDR)**, scaled to millions for readability.

üß† This trend may inform decisions on marketing spend, inventory planning, or pricing strategy.

### üì¶ Monthly Quantity Trend

```python
# Set Date as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Compute monthly quantity sold
monthly_qty = df['Qty'].resample('M').sum()
monthly_qty = monthly_qty[monthly_qty.index.month != 3]  # Remove March
monthly_qty.index = monthly_qty.index.strftime('%b %Y')

# Plot
plt.figure(figsize=(10, 5))
monthly_qty.plot(marker='o', color='orange', label='Total Quantity Sold')

# Titles and Labels
plt.title("Monthly Quantity Trend", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Reset Index
df.reset_index(inplace=True)

# Save as image in visualizations folder
plt.savefig("../visualizations/monthly_quantity.png", dpi=300)

plt.show()
```

This line chart displays the total quantity of items sold monthly between April and June.

The trend closely mirrors the revenue pattern observed earlier:

- **April** recorded the **highest quantity sold**, aligning with the month of highest revenue.
- **June** had the **lowest sales volume**, reinforcing the downward trend.
- This parallel suggests that fluctuations in revenue are largely driven by changes in order volume rather than price.

üìâ Monitoring quantity alongside revenue can help pinpoint whether performance shifts are volume-related or pricing-related.

### üìâ Daily Revenue vs Quantity Sold

```python
# Set Date as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Resample daily totals
daily_revenue = df['Amount (IDR)'].resample('D').sum() / 1_000_000  # Convert to millions
daily_qty = df['Qty'].resample('D').sum()

# Plot with two y-axes
fig, ax1 = plt.subplots(figsize=(14, 6))

# Plot revenue
ax1.plot(daily_revenue, color='teal')
ax1.set_ylabel("Revenue (Million IDR)", color='teal')
ax1.tick_params(axis='y', labelcolor='teal')

# Format x-axis
import matplotlib.dates as mdates
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax1.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
plt.xticks(rotation=45)

# Plot quantity on secondary y-axis
ax2 = ax1.twinx()
ax2.plot(daily_qty, color='orange')
ax2.set_ylabel("Quantity Sold", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Title and Layout
plt.title("Daily Revenue vs Quantity Sold", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()

# Reset index
df.reset_index(inplace=True)

# Save as image
plt.savefig("../visualizations/daily_revenue_quantity.png", dpi=300)

plt.show()
```

This dual-axis line chart provides a day-by-day comparison between total revenue and quantity sold.

- **Teal line:** Daily revenue (in millions of IDR).
- **Orange line:** Total quantity sold per day.
- The two metrics closely follow each other, suggesting that sales volume is the main driver of revenue changes.

üß† This view is especially helpful to detect short-term spikes or drops in activity that monthly aggregations may hide. It can assist with identifying:
- The impact of promotions or events.
- Drops in operational performance or stock-outs.
- Shifts in customer behavior over time.


üìâ **Smoothed Trend ‚Äì Daily Revenue vs Quantity Sold (7-Day Rolling Average)**

While the original chart accurately reflects real daily fluctuations, it can sometimes be difficult to interpret the overall direction due to short-term noise.

This smoothed version uses a **7-day rolling average** to:
- Highlight **underlying trends** more clearly.
- Make patterns easier to read, especially for non-technical audiences.
- Reduce distractions from isolated outliers or daily variance.

Such smoothing is useful when communicating broader trends, making the graph a strong addition to dashboards or executive summaries.


```python
# Set Date as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Resample daily totals
daily_revenue = df['Amount (IDR)'].resample('D').sum() / 1_000_000  # Convert to millions
daily_qty = df['Qty'].resample('D').sum()

# Apply 7-day rolling average
smoothed_revenue = daily_revenue.rolling(window=7, center=True).mean()
smoothed_qty = daily_qty.rolling(window=7, center=True).mean()

# Plot
fig, ax1 = plt.subplots(figsize=(14, 6))

# Smoothed revenue line
ax1.plot(smoothed_revenue, color='teal')
ax1.set_ylabel("Revenue (Million IDR)", color='teal')
ax1.tick_params(axis='y', labelcolor='teal')

# Format x-axis
import matplotlib.dates as mdates
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax1.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
plt.xticks(rotation=45)

# Smoothed quantity line
ax2 = ax1.twinx()
ax2.plot(smoothed_qty, color='orange')
ax2.set_ylabel("Quantity Sold", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Final layout
plt.title("Smoothed Daily Revenue vs Quantity Sold (7-Day Rolling Average)")
plt.tight_layout()
plt.grid(True)

# Save image
plt.savefig("../visualizations/smoothed_daily_revenue_quantity.png", dpi=300)

plt.show()

# Reset index
df.reset_index(inplace=True)
```

üìä **Pricing Insight from Smoothed Trends**

Notice how the **revenue** (teal) line occasionally diverges from the **quantity sold** (orange) line:

- When the **revenue curve is above** the quantity curve ‚Üí **higher-value items** are being sold.
- When the **quantity curve is higher** ‚Üí more **lower-priced items** are being purchased in greater volume.

These divergences can reflect:
- Promotional events or discount periods,
- Shifts in consumer preferences toward different product tiers,
- Introduction of new premium or budget products.

Understanding this pattern can help refine pricing strategies, forecast profitability, and adjust inventory to match consumer behavior.


### üßæ Average Order Value (Daily) with Overall Benchmark

```python
# Set Date as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Compute daily AOV: total revenue / unique orders
daily_stats = df.resample('D').agg({
    'Amount (IDR)': 'sum',
    'Order ID': 'nunique'
})
daily_stats['AOV'] = daily_stats['Amount (IDR)'] / daily_stats['Order ID']

# Compute overall AOV benchmark
total_revenue = df['Amount (IDR)'].sum()
total_orders = df['Order ID'].nunique()
overall_aov = total_revenue / total_orders

# Plot
plt.figure(figsize=(14, 6))
daily_stats['AOV'].plot(color='purple', label='Daily AOV')
plt.axhline(overall_aov, color='gray', linestyle='--', linewidth=2,
            label=f'Overall AOV ({overall_aov:.0f} IDR)')

# Titles and labels
plt.title("Average Order Value (Daily) with Overall Benchmark", fontsize=14)
plt.xlabel("Date")
plt.ylabel("AOV (IDR)")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save as image
plt.savefig("../visualizations/daily_aov_vs_benchmark.png", dpi=300)

# Show plot
plt.show()

# Reset index
df.reset_index(inplace=True)
```

This chart shows the **daily average order value (AOV)** and compares it to the **overall benchmark AOV** for the dataset.

- **AOV** is computed as: total daily revenue divided by the number of unique orders.
- The **benchmark line** provides a stable reference to spot deviations in customer purchase behavior.

üß† **Linked Insight (from: _Smoothed Daily Revenue vs Quantity Sold_)**  
In the visual titled **"Smoothed Daily Revenue vs Quantity Sold (7-Day Rolling Average)"**, we noticed periods where revenue and quantity trends diverged ‚Äî indicating shifts in item value rather than just volume.

This AOV chart confirms and quantifies that:
- When the **revenue curve exceeds** the quantity curve, the AOV tends to rise above the benchmark.
- When **quantity surpasses** revenue, the AOV falls ‚Äî signaling more orders of low-priced items.

üìà **Why this matters:**
Tracking AOV in context helps identify:
- Changes in product mix (e.g., shift to higher- or lower-priced items).
- Effects of promotions or discounts.
- Days with unusually valuable or unprofitable order patterns.

Together with the smoothed trends, this chart provides a richer understanding of sales dynamics.
